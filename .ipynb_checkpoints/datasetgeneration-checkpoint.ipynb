```python
import numpy as np
from numpy import random
import csv 

s = input("Enter initial s:")
print("size of data: ", s)
s = int(s)
data = []

with open("input/input.csv", 'w', newline='') as file:
    writer = csv.writer(file)
    
    # Program will generate datasets for 6 different sizes (s)
    for x in range(6):
        print("size of data: ", s)
        # generate a dataset of size s with 0% duplicates that is sorted in ascending order
        data = list(range(1, s+1))
    
        print("Ordered array with 0% duplicates", data)
        writer.writerow(data)
        
         # generate a dataset of size s with 0% duplicates such that 60% of the array is already sorted ascending
        def shuffler (data, percent): 
            for i in range(percent-1,0,-1):
                j = random.randint(0,i+1)
                data[i],data[j] = data[j],data[i]
            return data
        percent = int(s * .4)# will shuffle 40% of the array leaving the other 60% sorted ascending
        print("60% ordered array: ",shuffler(data, percent))
        writer.writerow(data)
        
        # generate a randomized dataset of size s with 0% duplicates
        random.shuffle(data)
        print("Randomized dataset: ", data)
        writer.writerow(data)
        
       

    #generate a randomized dataset of size s with 20% and 40% duplicates 
        def percentDupe(size, percent):
            data = list(range(1, int(size*percent/100)+1))
            data.extend(list(range(1, int(size*(100-percent)/100)+1)))
            if len(data) > size:
                data.pop();
            
            return data
        
        data = percentDupe(s, 20)
        print("Data size: ",len(data))
        random.shuffle(data)
        print("Dataset with 20% duplicates", data)
        writer.writerow(data)
        
        data = percentDupe(s, 40)
        print("Data size: ",len(data))
        random.shuffle(data)
        print("Dataset with 40% duplicates", data)
        writer.writerow(data)
        
   # Increment the dataset size for the next iteration
        s = s*5

    
```
